### VIF

```python
x = df.drop(columns = ['Type of glass'])
y = df['Type of glass']

vif = pd.DataFrame()

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif["VIF"] = [variance_inflation_factor (x.values, i) for i in range(x.shape[1])]
vif["variables"] = x.columns
vif
```

### Contribution Score

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
x = df.drop(columns = ['Type of glass'])
y = df['Type of glass']
```

```python
best_features = SelectKBest(score_func=f_classif, k=9)

fit = best_features.fit(x,y)

df_scores = pd.DataFrame(fit.scores_) 
df_columns = pd.DataFrame(x.columns) 

feature_scores = pd.concat([df_columns, df_scores], axis=1)

feature_scores.columns = ['Feature_Name', 'Score']
print(feature_scores.nlargest(9, 'Score'))
```

# 4) Visual Representation

## Checking Skewness

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20,15), facecolor='red')
plotnumber = 1

for column in df:                       
    if plotnumber<=9:                    
        ax = plt.subplot(3,3,plotnumber)    
        sns.distplot(df[column])       
        plt.xlabel(column, fontsize=20)   

    plotnumber+=1  
plt.tight_layout()  
```

```python
df_features = df.drop(columns = ['Type of glass'])
```

```python
plt.figure(figsize=(20,25))
graph = 1

for column in df_features: 
    if graph<=9:
        plt.subplot(3,3, graph)
        ax=sns.boxplot(data= df_features [column])   
        plt.xlabel(column, fontsize=15)
    
    graph+=1
plt.show()
```

# 6) Selection of Models

### Creating Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
```

### Separating Features & Target

```python
x = df.drop(columns = ['Type of glass'])
y = df['Type of glass']
```

### Importing important Metrices

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score, classification_report
from sklearn.model_selection import GridSearchCV
```

### 

### 

### 

# 7) Standardization 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

### 

### 

# 8) Test - Train Split 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled, y, test_size = 0.25)
```

### 

### 

# 9) Training the Model

## Logistic Regression

### Creating Logistic Regression model

```python
cls_reg = LogisticRegression()
```

```python
cls_reg.fit(x_train, y_train)
```

```python
cls_reg.coef_
```

```python
cls_reg.intercept_
```

```python
print('Accuracy Score:',cls_reg.score(x_train, y_train))
```

```python
y_pred = cls_reg.predict(x_test)
y_pred
```

```python
print('Accuracy Score:',cls_reg.score(x_test, y_test))
```

### Confusion Matrix

```python
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```

### Classification Report

```python
print(classification_report(y_test,y_pred))
```

# 

# 

## DecisionTreeClassifier

### Creating DecisionTreeClassifier model

```python
dc = DecisionTreeClassifier()
param_grid = {'criterion': ['gini', 'entropy'],
              'splitter': ['best', 'random'],
              'max_depth': range(4, 20, 2),
              'min_samples_split': range(0, 10, 2),
              'min_samples_leaf': range(0,5,1),
              'max_features': ['sqrt', 'log2', None]
             }
gridsearch = GridSearchCV(estimator=dc, param_grid=param_grid)
gridsearch.fit(x_train, y_train)
gridsearch.best_params_
```

```python
cls_dec = DecisionTreeClassifier(criterion='entropy',min_samples_leaf= 2, min_samples_split=2,max_depth=16)
```

```python
cls_dec.fit(x_train, y_train)
```

```python
print('Accuracy Score:',cls_dec.score(x_train, y_train))
```

```python
y_pred = cls_dec.predict(x_test)
y_pred
```

```python
print('Accuracy Score:',cls_dec.score(x_test, y_test))
```

### Confusion Matrix

```python
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```

### Classification Report

```python
print(classification_report(y_test,y_pred))
```
